```python
import requests
import bs4

# load and get the website
response = requests.get('https://www.empik.com/bestsellery')

# create the soup
soup = bs4.BeautifulSoup(response.text, "html.parser")

# find all the tags with class city or number
data = soup.findAll(attrs={'class':['menu-hot-phrase','fips','tot-pop','median-income','no-housing-units','med-home-val','owner-occupied','house-w-debt','house-wo-debt']})
data = soup.findAll(attrs={'class':'ta-product-title'})
# print 'data' to console
print(data)
```
